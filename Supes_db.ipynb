```python
#update eye color
for doc in super_db.supestats.find({'appearance.eyeColor': "Bown"}):
    eyeresult = super_db.supestats.update_many({'appearance.eyeColor': 'Bown'}, {'$set':{'appearance.eyeColor': 'Brown'}})
    print(eyeresult.matched_count)
    print(eyeresult.modified_count)
```

```python
#update hair color
for doc in super_db.supestats.find({'appearance.hairColor': "Brownn"}):
    hairresult = super_db.supestats.update_many({'appearance.hairColor': 'Brownn'}, {'$set':{'appearance.hairColor': 'Brown'}})
    print(hairresult.matched_count)
    print(hairresult.modified_count)
```
